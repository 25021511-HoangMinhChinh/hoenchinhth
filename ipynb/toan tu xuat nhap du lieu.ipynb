```python
# W2A3 - ModAndDiv (không dùng input, sửa trực tiếp a, b)
a = 10
b = 3

tong = a + b
hieu = a - b
tich = a * b
phan_nguyen = a // b
phan_du = a % b
chia_thuc = a / b

print(tong)
print(hieu)
print(tich)
print(phan_nguyen)
print(phan_du)
print(f"{chia_thuc:.2f}")
```

```python
# W2A4 - AverageOfSix
# a1, b1, c1 hệ số 1; a2, b2 hệ số 2; a3 hệ số 3
a1, b1, c1, a2, b2, a3 = 7.0, 8.0, 9.0, 6.5, 7.5, 8.0  # sửa điểm nếu cần

tb = ((a1 + b1 + c1) + (a2 + b2) * 2 + a3 * 3) / 10
print(f"{tb:.1f}")
```

```python
# W2A6 - StringDataType
# Nhập một chữ thường 'a'..'z', in ra mã Unicode và chữ hoa tương ứng
c = 'a'  # sửa ký tự nếu cần, nhớ dùng chữ thường
code_point = ord(c)
upper_c = chr(code_point - 32)  # chỉ đúng với chữ cái latin thường

print(code_point)
print(upper_c)
```

```python
# W2A9 - ShoppingOnline
# x: giá đồng hồ (USD)
x = 100.0  # sửa giá nếu cần

# phí vận chuyển trung gian: 10 USD
shipping = 10.0

# thuế 30% + 10% trên giá trị chiếc đồng hồ
tax = x * 0.30 + x * 0.10

total = x + shipping + tax
print(f"{total:.2f}")
```
